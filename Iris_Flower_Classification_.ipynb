## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1 Sepal Length and Species

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** There is no significant difference in sepal length among the three iris species (Setosa, Versicolor, and Virginica).

**Alternative Hypothesis (H1):** There is a significant difference in sepal length among the three iris species.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats

# Perform one-way ANOVA
setosa_sepal_length = df[df['target'] == 0]['sepal length (cm)']
versicolor_sepal_length = df[df['target'] == 1]['sepal length (cm)']
virginica_sepal_length = df[df['target'] == 2]['sepal length (cm)']

f_statistic, p_value = stats.f_oneway(setosa_sepal_length, versicolor_sepal_length, virginica_sepal_length)
print("P-value for Sepal Length and Species:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

I Perform one-way ANOVA

##### Why did you choose the specific statistical test?

The purpose of this test is to determine whether there is a significant difference in sepal length among the three iris species (Setosa, Versicolor, and Virginica).
ANOVA is chosen because it is suitable for comparing the means of multiple groups (in this case, three species).
The null hypothesis (H0) assumes that there is no significant difference in sepal length among the species, while the alternative hypothesis (H1) suggests that there is a significant difference.

**P-Value:** The very small p-value (approximately 8.97e-24) indicates strong evidence against the null hypothesis. We reject the null hypothesis and conclude that there is a significant difference in sepal length among the three iris species.

### Hypothetical Statement - 2 Petal Length and Species

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** There is no significant difference in petal length among the three iris species (Setosa, Versicolor, and Virginica).

**Alternative Hypothesis (H1):** There is a significant difference in petal length among the three iris species.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
# Perform one-way ANOVA
setosa_petal_length = df[df['target'] == 0]['petal length (cm)']
versicolor_petal_length = df[df['target'] == 1]['petal length (cm)']
virginica_petal_length = df[df['target'] == 2]['petal length (cm)']

f_statistic, p_value = stats.f_oneway(setosa_petal_length, versicolor_petal_length, virginica_petal_length)
print("P-value for Petal Length and Species:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

##### Why did you choose the specific statistical test?

This test aims to determine whether there is a significant difference in petal length among the three iris species (Setosa, Versicolor, and Virginica).

The null hypothesis (H0) assumes that there is no significant difference in petal length among the species, while the alternative hypothesis (H1) suggests that there is a significant difference.

```python
# Identify outliers using the IQR method
def identify_outliers(data, feature_name):
    Q1 = data[feature_name].quantile(0.25)
    Q3 = data[feature_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[feature_name] < lower_bound) | (data[feature_name] > upper_bound)]
    return outliers
```

```python
# Identify outliers for Sepal Length (cm)
sepal_length_outliers = identify_outliers(df, 'sepal length (cm)')
print("Outliers for Sepal Length (cm):")
print(sepal_length_outliers)
```

### ML Model - 1

```python
# ML Model - 1 Implementation

# Split the data into training and testing sets
X = df[['sepal length (cm)', 'petal length (cm)']]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the SVM classifier
svm_classifier = SVC(kernel='linear', C=1)

# Train the classifier on the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", confusion)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python

# Define the grid for plotting the decision boundary
h = 0.02  # Step size in the mesh
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Make predictions on the grid
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='target', data=df)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.title("SVM Decision Boundary")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

### ML Model - 2

```python
from sklearn.ensemble import RandomForestClassifier
# Split the data into training and testing sets
X = df[['sepal length (cm)', 'petal length (cm)']]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)

# Print the results
print("Random Forest Classifier Results:")
print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", confusion)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
# Visualize the decision boundary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='target', data=df)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 200), np.linspace(ylim[0], ylim[1], 200))
Z = rf_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.title("Random Forest Decision Boundary")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
